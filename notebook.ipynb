```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

We need to Subset the DataFrame for the type "Movie" to keep only the movies. Also, we have to filter movies that was released between the year 1990 to 1999

```python
print(netflix_df.head())
```

```python
movies_90s = netflix_df[(netflix_df["type"] == "Movie") & (netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999)]
```

Now to find out the most frequent movie duration in the 90's era, we will use **mode()** method.

```python
s = pd.Series(movies_90s["duration"])
duration = s.mode()[0]
print(f"Most frequent movie duration in the 1990s were {duration}")
```

```python
# Plotting graph for a visual representation of result
plt.hist(movies_90s["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

For the second assignment, we need to complete 2 tasks
1. Filter out all "action movies" genre from movie 
2. Figure out movies that has duration lesser than 90 minutes

```python
# Filter out all 90s movies from "Action" genre
action_movies = movies_90s[movies_90s["genre"] == "Action"]

# Count all action movies that lasted less than 90 minutes
short_movie_count = 0

for label, row in action_movies.iterrows():
    if row["duration"] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count
print(f"A total of {short_movie_count} short action movies were released in the 1990s")
```
